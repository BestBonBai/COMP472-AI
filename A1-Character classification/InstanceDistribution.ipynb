```python
import config
import ProcessDataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sort the distribution dict according to the order of the character 
def sortCounts(counts, info_dict):
    result = list()
    for number in info_dict.keys():
        result.append(counts[number])
    return result

def plotDist(counts, letters, title):
    fig = plt.figure(figsize=(10,5))
    ax = fig.add_axes([0,0,1,1])
    ax.bar(letters, counts)
    plt.ylabel('Instance Occurences')
    plt.xlabel('Letters')
    plt.title(title)
    plt.show()
```

# Instance Distribution of Latin letters Data Set

```python
letters = config.LATIN_LETTERS
```

### Distribution of Latin letters in the training data set

```python
counts = ProcessDataset.countDistribution(config.LATIN_TRAIN_SET)
counts = sortCounts(counts, config.LATIN_INFO_DICT)
plotDist(counts, letters, "Latin letters distribution in training data set")
```

### Distribution of Latin letters in the testing data set with label

```python
counts.clear
counts = ProcessDataset.countDistribution(config.LATIN_TEST_LABEL)
counts = sortCounts(counts, config.LATIN_INFO_DICT)
plotDist(counts, letters, "Latin letters distribution in testing data set with label")
```
