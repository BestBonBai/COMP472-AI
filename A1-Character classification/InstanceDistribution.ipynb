```python
from utils import config, ProcessDataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sort the distribution dict according to the order of the character 
def sortCounts(counts, info_dict):
    result = list()
    for number in info_dict.keys():
        result.append(counts[number])
    return result

# Plot the instance distribution
def plotDist(counts, letters, title):
    fig = plt.figure(figsize=(10,5))
    ax = fig.add_axes([0,0,1,1])
    ax.bar(letters, counts)
    plt.ylabel('Instance Occurences')
    plt.xlabel('Letters')
    plt.title(title)
    plt.show()

# Show the character example in the dataset
def preview_char(data, letters, title, isGreekData):
    fig = plt.figure(figsize=(10,10))
    data = data.set_index("CHARACTER", drop=False)
    plt.title(title)
    for i in range(len(letters)):
        label = i
        pixels = data[data['CHARACTER'] == i].iloc[0]
        pixels = np.array(pixels[:-1], dtype='uint8')
        pixels = pixels.reshape((32, 32))
        fig.add_subplot(2,5,i+1) if (isGreekData) else fig.add_subplot(5,6,i+1)
        plt.imshow(pixels, cmap='gray')
```

# Instance Distribution of Latin letters Data Set

```python
letters = config.LATIN_LETTERS
latin_data = ProcessDataset.get_DataFromCSV(config.LATIN_TRAIN_SET)
preview_char(latin_data, letters, 'Latin Character Example', isGreekData=False)
```

### Distribution of Latin letters in the training data set

```python
counts = ProcessDataset.countDistribution(config.LATIN_TRAIN_SET)
counts = sortCounts(counts, config.LATIN_INFO_DICT)
plotDist(counts, letters, "Latin letters distribution in training data set")
```

### Distribution of Latin letters in the testing data set with label

```python
counts.clear
counts = ProcessDataset.countDistribution(config.LATIN_TEST_LABEL)
counts = sortCounts(counts, config.LATIN_INFO_DICT)
plotDist(counts, letters, "Latin letters distribution in testing data set with label")
```

### Distribution of Latin letters in the validation data set

```python
counts.clear
counts = ProcessDataset.countDistribution(config.LATIN_DATA_VAL)
counts = sortCounts(counts, config.LATIN_INFO_DICT)
plotDist(counts, letters, "Latin letters distribution in validation data set")
```
